```python
def process_dataframe(df):
    """Drop non-numeric columns and compute the mean of the remaining columns."""
    df_numeric = df.drop(columns=['filename'])  # Menghapus kolom 'filename' jika ada
    return df_numeric.mean()  # Mengembalikan rata-rata setiap kolom fitur

# Proses semua dataframe kain
mean_kh = process_dataframe(df_kh)  # Katun Hitam
mean_kp = process_dataframe(df_kp)  # Katun Putih
mean_ph = process_dataframe(df_ph)  # Polyester Hitam
mean_pp = process_dataframe(df_pp)  # Polyester Putih

# Gabungkan semua rata-rata ke dalam satu dataframe
mean_df = pd.DataFrame({
    'Katun Hitam': mean_kh,
    'Katun Putih': mean_kp,
    'Polyester Hitam': mean_ph,
    'Polyester Putih': mean_pp
})

# Bulatkan angka ke 4 angka di belakang koma
mean_df = mean_df.round(6)

# Ubah tampilan agar tidak dalam format ilmiah
pd.set_option('display.float_format', '{:.6f}'.format)

# Tampilkan hasil akhir
mean_df
```

```python
# Fungsi untuk mengambil standar deviasi dari setiap fitur numerik
def process_dataframe2(df):
    """Drop non-numeric columns and compute the standard deviation of the remaining columns."""
    df_numeric = df.drop(columns=['filename'])  # Menghapus kolom non-numerik
    return df_numeric.std()  # Mengembalikan standar deviasi tiap kolom

# Proses semua dataframe kain
std_kh = process_dataframe2(df_kh)  # Katun Hitam
std_kp = process_dataframe2(df_kp)  # Katun Putih
std_ph = process_dataframe2(df_ph)  # Polyester Hitam
std_pp = process_dataframe2(df_pp)  # Polyester Putih

# Gabungkan semua standar deviasi ke dalam satu dataframe
std_df = pd.DataFrame({
    'Katun Hitam': std_kh,
    'Katun Putih': std_kp,
    'Polyester Hitam': std_ph,
    'Polyester Putih': std_pp
})

std_df
```

```python
import os
import matplotlib.pyplot as plt
import numpy as np

# Fungsi untuk membuat plot
def create_contrast_plot(fabrics, x_labels, tick_positions, contrast_data, title, output_path):
    # Membuat figure dengan ukuran yang lebih besar
    plt.figure(figsize=(14, 6))
    
    # Menggambar garis dan marker
    plt.plot(tick_positions, contrast_data, label='Contrast', marker='o', markersize=8, color='blue')
    
    # Menambahkan grid dengan opasitas rendah
    plt.grid(True, linestyle='--', alpha=0.6)
    
    # Menambahkan judul dan label sumbu
    plt.title(title, fontsize=14, fontweight='bold')
    plt.ylabel('Average Values', fontsize=12)
    plt.xticks(tick_positions, x_labels, rotation=0, fontsize=10)
    
    # Tambahkan label jenis kain di tengah setiap grup RGB
    for i, fabric in enumerate(fabrics):
        plt.text(i * n_channels + 1, plt.ylim()[1] - 0.0001, fabric, ha='center', va='top', fontsize=12, fontweight='bold')
    
    # Tambahkan garis pemisah antar jenis kain
    for i in range(1, len(fabrics)):
        plt.axvline(x=i * n_channels - 0.5, color='black', linestyle='--', lw=1)
    
    # Menambahkan legenda dengan posisi yang lebih baik
    plt.legend(loc='upper right', fontsize=10)
    
    # Mengatur layout agar elemen-elemen tidak saling tumpang tindih
    plt.tight_layout()
    
    # Simpan plot ke dalam file
    plt.savefig(output_path, dpi=400)
    
    # Tampilkan plot
    plt.show()

# Define folder output (pastikan folder sudah ada)
output_folder = r"D:\SEMESTER6\PengenalaPola\analisis_kain_glcm_rgb\outputs\lbp_plots_analysis"

# Plot untuk kain warna hitam
file_name_black = "contrast_jenis_kain_hitam.png"
output_path_black = os.path.join(output_folder, file_name_black)
create_contrast_plot(
    fabrics_black,
    x_labels_black,
    tick_positions_black,
    contrast_black,
    "Contrast pada Jenis Kain Warna Hitam di Setiap Channel RGB dengan LBP",
    output_path_black
)

# Plot untuk kain warna putih
file_name_white = "contrast_jenis_kain_putih.png"
output_path_white = os.path.join(output_folder, file_name_white)
create_contrast_plot(
    fabrics_white,
    x_labels_white,
    tick_positions_white,
    contrast_white,
    "Contrast pada Jenis Kain Warna Putih di Setiap Channel RGB dengan LBP",
    output_path_white
)
```

```python
import pandas as pd
```

```python
# Load the CSVs for different percentages (if available)
path_katun_hitam = r"D:\SEMESTER6\PengenalaPola\analisis_kain_glcm_rgb\outputs\glcm_results\katun_htm_glcm_result.csv"
path_katun_putih = r"D:\SEMESTER6\PengenalaPola\analisis_kain_glcm_rgb\outputs\glcm_results\katun_pth_glcm_result.csv"
path_polyester_hitam = r"D:\SEMESTER6\PengenalaPola\analisis_kain_glcm_rgb\outputs\glcm_results\polyester_htm_glcm_result.csv"
path_polyester_putih = r"D:\SEMESTER6\PengenalaPola\analisis_kain_glcm_rgb\outputs\glcm_results\polyester_pth_glcm_result.csv"


# Reading the data assuming they follow the same structure
df_kh = pd.read_csv(path_katun_hitam)
df_kp = pd.read_csv(path_katun_putih)
df_ph = pd.read_csv(path_polyester_hitam)
df_pp = pd.read_csv(path_polyester_putih)
```

```python
# process the data transform
def process_dataframe(df):
    """Drop non-numeric columns and compute the mean of the remaining columns."""
    df_numeric = df.drop(columns=['filename'])  # Dropping the 'filename' column
    return df_numeric # Returning the mean of each column


# Process each dataframe
df_kh = process_dataframe(df_kh)
df_kp = process_dataframe(df_kp)
df_ph = process_dataframe(df_ph)
df_pp = process_dataframe(df_pp)
# Combine all the means into a single dataframe
df_combined = pd.concat([df_kh, df_kp, df_ph, df_pp], axis=0, keys=['katun_htm', 'katun_pth', 'polyester_htm', 'polyester_pth'])

df_combined['class'] = df_combined.index.get_level_values(0)

df_combined.info()
```
